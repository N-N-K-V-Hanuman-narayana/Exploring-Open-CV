```python
# Import required packages

import cv2
import matplotlib.pyplot as plt

# Read image
image_Original = cv2.imread("mani2.jpg")
image_MedianBlur = cv2.imread("mani2.jpg")
image_GaussianBlur = cv2.imread("mani2.jpg")
image_BilateralBlur = cv2.imread("mani2.jpg")

# Apply Median Blur
# Removes salt-and-pepper noise
image_MedianBlur = cv2.medianBlur(image_MedianBlur, 9)

# Apply Gaussian Blur
# Smooths image using Gaussian kernel
image_GaussianBlur = cv2.GaussianBlur(
    image_GaussianBlur,
    (9,9),
    10
)

# Apply Bilateral Filter
# Smooths image while preserving edges
image_BilateralBlur = cv2.bilateralFilter(
    image_BilateralBlur,
    9,
    100,
    75
)

# Convert BGR to RGB for display

image_Original = cv2.cvtColor(image_Original, cv2.COLOR_BGR2RGB)
image_MedianBlur = cv2.cvtColor(image_MedianBlur, cv2.COLOR_BGR2RGB)
image_GaussianBlur = cv2.cvtColor(image_GaussianBlur, cv2.COLOR_BGR2RGB)
image_BilateralBlur = cv2.cvtColor(image_BilateralBlur, cv2.COLOR_BGR2RGB)

# Display images

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(image_Original)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(image_MedianBlur)
plt.title("Median Blur")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(image_GaussianBlur)
plt.title("Gaussian Blur")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(image_BilateralBlur)
plt.title("Bilateral Filter")
plt.axis("off")

plt.show()
```
