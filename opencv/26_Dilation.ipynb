```python
# Erosion using different structuring elements

import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread("mani2.jpg")

# Erosion sizes
s1 = 0
s2 = 10
s3 = 10

# Structuring element types
t1 = cv2.MORPH_RECT      # Rectangle
t2 = cv2.MORPH_CROSS     # Cross
t3 = cv2.MORPH_ELLIPSE   # Ellipse

# Create kernels
tmp1 = cv2.getStructuringElement(
    t1,
    (2*s1 + 1, 2*s1 + 1),
    (s1, s1)
)

tmp2 = cv2.getStructuringElement(
    t2,
    (2*s2 + 1, 2*s2 + 1),
    (s2, s2)
)

tmp3 = cv2.getStructuringElement(
    t3,
    (2*s3 + 1, 2*s3 + 1),
    (s3, s3)
)

# Apply erosion
final1 = cv2.erode(image, tmp1)
final2 = cv2.erode(image, tmp2)
final3 = cv2.erode(image, tmp3)

# Convert BGR to RGB
final1 = cv2.cvtColor(final1, cv2.COLOR_BGR2RGB)
final2 = cv2.cvtColor(final2, cv2.COLOR_BGR2RGB)
final3 = cv2.cvtColor(final3, cv2.COLOR_BGR2RGB)

# Display outputs
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(final1)
plt.title("Rectangle Erosion")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(final2)
plt.title("Cross Erosion")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(final3)
plt.title("Ellipse Erosion")
plt.axis("off")

plt.show()
```
