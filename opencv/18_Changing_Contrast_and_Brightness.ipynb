```python
from skimage import io
import os

IMAGE_FILE = "mani2.jpg"  # change if needed

if not os.path.exists(IMAGE_FILE):
    raise FileNotFoundError(f"{IMAGE_FILE} not found in notebook folder")

from skimage import io
import os

IMAGE_FILE = "mani2.jpg"  # change if needed

if not os.path.exists(IMAGE_FILE):
    raise FileNotFoundError(f"{IMAGE_FILE} not found in notebook folder")

from skimage import io, transform

# Read image
img = io.imread(IMAGE_FILE)

# Affine transformation
tform = transform.AffineTransform(
    scale=(1.2, 1.2),
    rotation=0.3,
    translation=(50, 30)
)

affine_img = transform.warp(img, tform)

# Display image
io.imshow(affine_img)
io.show()
```

```python
from skimage import io
import os

IMAGE_FILE = "mani2.jpg"  # change if needed

if not os.path.exists(IMAGE_FILE):
    raise FileNotFoundError(f"{IMAGE_FILE} not found in notebook folder")

from skimage import io
import os

IMAGE_FILE = "mani2.jpg"  # change if needed

if not os.path.exists(IMAGE_FILE):
    raise FileNotFoundError(f"{IMAGE_FILE} not found in notebook folder")

from skimage import io, transform
import matplotlib.pyplot as plt

# Read image
img = io.imread(IMAGE_FILE)

# Rotate image
rotated = transform.rotate(img, 45)

# Display both images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(rotated)
plt.title("Rotated")

plt.show()
```

```python
from skimage import io
import os

IMAGE_FILE = "mani2.jpg"  # change if needed

if not os.path.exists(IMAGE_FILE):
    raise FileNotFoundError(f"{IMAGE_FILE} not found in notebook folder")

from skimage import io
import os

IMAGE_FILE = "mani2.jpg"  # change if needed

if not os.path.exists(IMAGE_FILE):
    raise FileNotFoundError(f"{IMAGE_FILE} not found in notebook folder")

from skimage import io, transform
import numpy as np

# Read image
img = io.imread(IMAGE_FILE)

# Rotate image
img = transform.rotate(img, 30)

# Resize image
img = transform.resize(img, (400, 400))

# Flip horizontally
img = np.fliplr(img)

# Display final image
io.imshow(img)
io.show()
```
