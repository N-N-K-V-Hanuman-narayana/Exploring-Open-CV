```python
import cv2
import numpy as np

img = np.ones((500, 500, 3), dtype=np.uint8) * 255

p1 = (100, 100)
p2 = (300, 150)
p3 = (200, 350)

cv2.circle(img, p1, 5, (0, 0, 255), -1)
cv2.circle(img, p2, 5, (0, 0, 255), -1)
cv2.circle(img, p3, 5, (0, 0, 255), -1)

cv2.line(img, p1, p2, (255, 0, 0), 2)
cv2.line(img, p2, p3, (255, 0, 0), 2)
cv2.line(img, p3, p1, (255, 0, 0), 2)

cv2.imwrite("triangle.jpg", img)   # Save image

cv2.imshow("Points and Lines", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread("mani2.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("Original", img)
cv2.imshow("Black and White", gray)

cv2.imwrite("mani2_bw.jpg", gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("mani2.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("mani2.jpg")

# Convert BGR to RGB for correct colors
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

# Black & White Image
plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.show()
```

```python
import cv2 
import matplotlib.pyplot as plt

img = cv2.imread("mani2.jpg")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize to 400x400 pixels
resized = cv2.resize(img_rgb, (800, 800))

plt.imshow(resized)
plt.title("Resized Image")
plt.axis("on")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((400, 400, 3), dtype=np.uint8)

cv2.line(
    img,
    (50, 50),
    (350, 350),
    (0, 255, 0),  # Green (BGR)
    5
)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = np.zeros((400,400,3), dtype=np.uint8)

cv2.rectangle(
    img,
    (100,100),      # top-left
    (300,250),      # bottom-right
    (255,0,0),      # blue (BGR)
    3
)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
